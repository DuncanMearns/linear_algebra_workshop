# 1. Data structures

## Python built-ins

### Lists

```python
my_list = ["a", "b", "c"]
print(my_list)
```

List operations:
- assignment (mutates)
- append
- pop

```python
my_list[0] = "d"
print(my_list)
```

```python
my_list.append("e")
print(my_list)
```

```python
my_list.pop(1)
print(my_list)
```

Lists of lists

```python
list_of_lists = [my_list, [1, 2, 3], [0.5]]
print(list_of_lists)
```

### Tuples

```python
my_tuple = ("a", "b", "c")
print(my_tuple)
```

Not allowed!

```python
my_tuple.append("e")
```

```python
my_tuple.pop(1)
```

### Range

```python
range(10)  # returns a range object
```

```python
list(range(10))  # range starts at zero, ends at n-1
```

```python
list(range(5, 10))  # first number to n-1
```

```python
list(range(5, 15, 3))  # first number to n-1 with step
```

## Iteration

```python
for i in range(10):
    print(i)
```

#### zip/enumerate

```python
first_names = ["Tom", "Dick", "Harry"]
last_names = ["Holland", "Cheney", "Potter"]

for first, last in zip(first_names, last_names):
    print(first, last)
```

```python
# "Non-pythonic"
indices = range(3)
for idx, name in zip(indices, first_names):
    print(idx, name)
```

```python
# "Pythonic"
for idx, name in enumerate(first_names):
    print(idx, name)
```

# Vectors

```python
v = [0.5, 1.]
```

```python
fig, ax = plt.subplots(figsize=(3, 3))
ax.quiver([0], [0], [v[0]], [v[1]], units="xy", scale=1)
ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
```

### Dimensionality

```python
v = [-0.5, 2., 1.5]
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter([0], [0], [0], c="k")
ax.quiver([0], [0], [0], [v[0]], [v[1]], [v[2]], color="k")
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
```
