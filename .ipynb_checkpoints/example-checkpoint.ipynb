```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.dpi'] = 150.
plt.rcParams['figure.figsize'] = (3, 3)
```

# Example data

```python
points = np.load("example_tracking.npy")
print(points.shape)
```

```python
tracking = pd.read_csv("example_tracking.csv")
heading = tracking.loc[:, "heading"].values
```

```python
colors = plt.colormaps["viridis"](np.linspace(0, 1, len(points)))
```

```python
fig, ax = plt.subplots()
ax.scatter(*points[:, 0].T, c=colors, s=1)
ax.axis("equal");
```

```python
fig, ax = plt.subplots()
ax.plot(heading)
```

Plot tail points (first 100 frames only)

```python
fig, ax = plt.subplots()
for i, ps in enumerate(points[:100]):
    ax.plot(*ps.T, lw=0.5)
ax.axis("equal");
```

Take the first bout

```python
bout1 = points[:100]
print(bout1.shape)  # time, tail_points, xy
```

Make head-centered

```python
centered = bout1 - bout1[:, [0], :]
```

```python
fig, ax = plt.subplots()
for i, ps in enumerate(centered):
    ax.plot(*ps.T, lw=0.5)
ax.axis("equal");
```

Compute rotation matrices

```python
thetas = heading[:100]
Rs = np.array([[np.cos(thetas), np.sin(thetas)],
               [-np.sin(thetas), np.cos(thetas)]])
Rs = np.rollaxis(Rs, 2)
print(Rs.shape)
```

```python
centered.shape
```

Rotate points (the slow way!)

```python
rotated = []
for i, ps in enumerate(centered):
    rot = Rs[i]
    ps_trans = []
    for xy in ps:
        xy_trans = np.dot(rot, xy)
        ps_trans.append(xy_trans)
    rotated.append(ps_trans)
rotated = np.array(rotated)
rotated.shape
```

Rotate points (the cool way!)

```python
rotated_einsum = np.einsum("tij,tnj->tni", Rs, centered)
print(rotated_einsum.shape)
```

Plot results

```python
fig, axes = plt.subplots(1, 2, figsize=(6, 3))

for i, ps in enumerate(rotated):
    axes[0].plot(*ps.T, lw=0.5)
axes[0].set_title("Slow")

for i, ps in enumerate(rotated_einsum):
    axes[1].plot(*ps.T, lw=0.5)
axes[1].set_title("Einsum")


for ax in axes:
    ax.axis("equal");
```
