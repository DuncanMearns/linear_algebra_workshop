```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
```

```python
%matplotlib widget
```

# Intro

```python
v1 = np.array([5, 7])
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

ax.quiver([0], [0], [v1[0]], [v1[1]], units="xy", angles="xy", scale_units="xy", scale=1, zorder=2)

ax.grid(visible=True, zorder=1)
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
ax.set_xticks(np.arange(9))
ax.set_yticks(np.arange(9));
```

## Unit vectors 

```python
i = np.array([1., 0.])
j = np.array([0., 1.])
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

for v in (i, j):
    ax.quiver([0], [0], [v[0]], [v[1]], units="xy", angles="xy", scale_units="xy", scale=1, zorder=2)
ax.scatter([0], [0], c="k", zorder=2)

ax.grid(visible=True, zorder=1)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_xticks(np.arange(-3, 4))
ax.set_yticks(np.arange(-3, 4));
```

```python
v1 = 5 * i + 7 * j
```

```python
v1
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

ax.quiver([0], [0], [v1[0]], [v1[1]], units="xy", angles="xy", scale_units="xy", scale=1, zorder=2)

ax.grid(visible=True, zorder=1)
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
ax.set_xticks(np.arange(9))
ax.set_yticks(np.arange(9));
```

## Vectors have a magnitude (length)

```python
# Pythagoras theorem
```

```python
length = np.sqrt(v1[0] ** 2 + v1[1] ** 2)
print(length)
```

```python
length = np.linalg.norm(v1)
print(length)
```

```python
v_normed = v1 / length
print(v_normed)
```

# Dot product

```python
v2 = np.array([12, -3])
```

```python
fig, ax = plt.subplots(figsize=(3, 3))
for v in (v1, v2):
    ax.quiver([0], [0], [v[0]], [v[1]], units="xy", angles="xy", scale_units="xy", scale=1, zorder=2)
ax.scatter([0], [0], c="k", zorder=2)

ax.grid(visible=True, zorder=1)
ax.set_xlim(-5, 15)
ax.set_ylim(-5, 15)
ax.set_xticks(np.arange(-5, 15))
ax.set_yticks(np.arange(-5, 15));
```

### Adding vectors

```python
v3 = v1 + v2
print(v3)
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

ax.plot([0, v1[0]], [0, v1[1]], c="b")
ax.plot([v1[0], v3[0]], [v1[1], v3[1]], c="r")
ax.quiver([0], [0], [v3[0]], [v3[1]], units="xy", angles="xy", scale_units="xy", scale=1, zorder=2)
ax.scatter([0], [0], c="k", zorder=2)

ax.grid(visible=True, zorder=1)
ax.set_xlim(-5, 20)
ax.set_ylim(-5, 20)
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(base=1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(base=5))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(base=1))
```

## Multiplying vectors (dot product)

```python
# Multiply xs and ys
product = v1 * v2
print(product)
```

```python
# then sum
np.sum(product)
```

```python
# Dot product
dot = np.dot(v1, v2)
print(dot)
```

### Why?

```python
print(v1, v2)
```

```python
v1_normed = v1 / np.linalg.norm(v1)
v2_normed = v2 / np.linalg.norm(v2)
```

```python
print(v1_normed, v2_normed)
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

ax.quiver([0], [0], [v1_normed[0]], [v1_normed[1]], units="xy", angles="xy", scale_units="xy", scale=1, zorder=2)
ax.quiver([0], [0], [v2_normed[0]], [v2_normed[1]], units="xy", angles="xy", scale_units="xy", scale=1, zorder=2)
ax.scatter([0], [0], c="k", zorder=2)

ax.grid(visible=True, zorder=1)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=.2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(base=.1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(base=.2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(base=.1))
```

```python
normed_dot = np.dot(v1_normed, v2_normed)
normed_dot
```

```python
v1_scaled = v1_normed * normed_dot
v2_scaled = v2_normed * normed_dot
```

```python
kwargs = dict(units="xy", angles="xy", scale_units="xy", scale=1, zorder=2)
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

ax.quiver([0], [0], [v1_normed[0]], [v1_normed[1]], **kwargs)
ax.quiver([0], [0], [v2_normed[0]], [v2_normed[1]], **kwargs)

ax.quiver([0], [0], [v1_scaled[0]], [v1_scaled[1]], color="r", **kwargs)
ax.quiver([0], [0], [v2_scaled[0]], [v2_scaled[1]], color="r", **kwargs)

ax.scatter([0], [0], c="k", zorder=2)

ax.grid(visible=True, zorder=1)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=.2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(base=.1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(base=.2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(base=.1))
```
